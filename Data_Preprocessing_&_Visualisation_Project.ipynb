```python
# plot a correlation map the database df 
def plot_correlation_map(df):

  # input the correlation for the database
  corr=df.corr()

  # Set up matplot
  s,ax=plt.subplots(figsize=(12,10))

  # set up diverging color map 
  cmap= sns.diverging_palette(250,15,as_cmap=True)

  # 
  s= sns.heatmap(corr,cmap=cmap, square=True, cbar_kws={'shrink':.9}, ax=ax, annot=True, annot_kws={"fontsize":12})

plot_correlation_map(df)
```

Correlation coefficient is used to check values which have a relation with each other. When the value between two variable is 1, that means one variable increases, the other one also increases. Note that when you have 1 as the correlation between two variables, its a perfect correlation and we have can see perfect correlations in the following variables.

familySize and relatives


 Also the closer the correlation is to 1 between two variables, that means both variables are correlated.

Example of such are: Fare_Per_Person and Fare which have a correlation of 0.84.
 Relatives and Parch which have a correlation of 0.78.

When the value is 0, there is no correlation. This simply means that the variables changes in a random manner with respect to each other.
From the map, we see that "Survived" does not correlate with any other variables
